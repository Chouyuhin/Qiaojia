## read denoised npz file

```python
import numpy as np

# load npz file
def load_npz_file(file_path):
    with np.load(file_path,allow_pickle=True) as data:
        data_array = data['data']
        noise_array = data['noise']
    return data_array, noise_array

# file example
file_path = 'g:/云南巧家宽频带数据/denoised_mseed/QJ01202208/results/segment_0.mseed.npz'  # 替换为你的文件路径
data, noise = load_npz_file(file_path)

print("Data:", data)
print("Noise:", noise)
print(data.shape)
```
