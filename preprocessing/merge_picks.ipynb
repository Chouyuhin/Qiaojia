## merge picks_results from all stations

```python
## 将多个picks文件合成为一个总picks文件，并存到gammaAssociation
import os
import pandas as pd

# 定义包含多个 CSV 文件的文件夹路径
output_file = "D:/MyRepository/Qiaojia/GammaAssociation/merged_picks.csv"  # 合并后的文件路径

# 获取文件夹中的所有 CSV 文件
files_path = "D:/MyRepository/all_denoised_model80"  # 文件夹路径
station_name = os.listdir(files_path) 
#print(station_name)
csv_files=[]
for station in station_name:
    for f in os.listdir(os.path.join(files_path,station)):
        if f.endswith(".csv"):
            csv_file = os.path.join(files_path,station, f) 
            csv_files.append(csv_file)
print(csv_files[:5])

# 定义统一的列顺序
expected_columns = ["station_id", "waveform_id", "phase_time", "phase_type", "phase_score", "phase_amplitude"]

# 初始化一个空的 DataFrame
merged_df = pd.DataFrame()
lenth=0
# 逐个读取并合并文件
for csv_file in csv_files:
    #print(csv_file)
    df = pd.read_csv(csv_file,index_col=False,header=1,names=expected_columns)  # 读取当前文件
    # 去掉列名空格并调整列顺序
    df.columns = df.columns.str.strip()
    df = df.dropna()
    #df = df[expected_columns]  # 调整列顺序
    # 修改 station_id 列，提取前四个字符
    lenth = lenth+len(df)
    print(len(df))
    df["station_id"] = df["station_id"].str[:4]  # 提取格式类似 QJ02 的部分
    print(df['station_id'][:5])
    merged_df = pd.concat([merged_df, df], ignore_index=True)  # 合并

# 将合并后的数据保存为新的 CSV 文件
merged_df.to_csv(output_file, index=False)
print(f"合并完成，保存到: {output_file}")
print(lenth)


```
